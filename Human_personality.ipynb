```python
import pandas as pd
```

```python
df=pd.read_csv('/content/personality_datasert.csv')
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Personality']=le.fit_transform(df['Personality'])
df['Drained_after_socializing']=le.fit_transform(df['Drained_after_socializing'])
df['Stage_fear']=le.fit_transform(df['Stage_fear'])
```

```python
df.dtypes
```

```python
df
```

```python
#slicing
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
#balancing
y.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
sm=SMOTE()
X_res,y_res=sm.fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_scaled=sc.fit_transform(X_res)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y_res,test_size=0.3,random_state=1)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
from sklearn.svm import SVC
svm=SVC()
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
from sklearn.metrics import classification_report
```

```python
lst=[lr,dt,rf,knn,svm,nb]
for i in lst:
  print("***********",i,"***********")
  i.fit(X_train,y_train)
  y_pred=i.predict(X_test)
  print(classification_report(y_test,y_pred))
```

```python
#hyper parametr tunning
from sklearn.model_selection import GridSearchCV
p={'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf', 'linear']}
g=GridSearchCV(svm,p,cv=10,scoring="accuracy")
g.fit(X_train,y_train)
```

```python
print(g.best_params_)
```

```python
svm1=SVC(C=0.1,gamma=0.1,kernel='rbf')
svm1.fit(X_train,y_train)
```

```python
y_pred=svm1.predict(X_test)
print(classification_report(y_test,y_pred))
```

```python
y_pred
```

```python
#predicting new values
y_new=svm1.predict([[4.0,0,4.0,6.0,0,13.0,5.0]])
print(y_new)
```

```python
import joblib
joblib.dump(svm1,'personality_model')
joblib.dump(sc,'scaler.joblib')
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import numpy as np
import joblib

st.set_page_config(page_title="PERSONALITY PREDICTION",page_icon="person",layout="wide")

st.title("🚶PERSONALITY PREDICTION🚶‍➡️")

# Load the model and scaler
svm_model = joblib.load('personality_model')
scaler = joblib.load('scaler.joblib')

st.header("Enter the following details to predict personality:")

# Create input fields for each feature
time_spent_alone = st.slider("Time spent alone (hours per week)", 0, 15, 5)
stage_fear = st.radio("Stage fear", ["No", "Yes"])
social_event_attendance = st.slider("Social event attendance (events per month)", 0, 15, 5)
going_outside = st.slider("Going outside (times per week)", 0, 15, 5)
drained_after_socializing = st.radio("Drained after socializing", ["No", "Yes"])
friends_circle_size = st.slider("Friends circle size", 0, 20, 10)
post_frequency = st.slider("Post frequency (posts per month)", 0, 15, 5)

# Map categorical inputs to numerical values
stage_fear_encoded = 1 if stage_fear == "Yes" else 0
drained_after_socializing_encoded = 1 if drained_after_socializing == "Yes" else 0

# Create a dataframe from the inputs
input_data = pd.DataFrame([[time_spent_alone, stage_fear_encoded, social_event_attendance, going_outside, drained_after_socializing_encoded, friends_circle_size, post_frequency]])

# Scale the input data
input_data_scaled = scaler.transform(input_data)

# Predict the personality
if st.button("Predict Personality"):
    prediction = svm_model.predict(input_data_scaled)

    if prediction[0] == 0:
        st.success("Predicted Personality: Extrovert")
    else:
        st.success("Predicted Personality: Introvert")
```
